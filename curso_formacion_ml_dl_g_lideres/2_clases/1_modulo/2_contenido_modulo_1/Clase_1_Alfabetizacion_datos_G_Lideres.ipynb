## 5. Actividad práctica guiada (Nivel básico)

En esta actividad trabajarás con un conjunto de datos sencillo y responderás preguntas estratégicas desde el rol de administrador(a).

```
import pandas as pd
import numpy as np

# Simulación de datos de ventas mensuales
datos = pd.DataFrame({
    'mes': ['Ene','Feb','Mar','Abr','May','Jun'],
    'ventas': [120, 150, 130, 180, 200, 220]
})

datos
```

### Análisis guiado

1. ¿En qué meses las ventas fueron mayores?
2. ¿Observas alguna tendencia general?
3. ¿Qué decisiones podría tomar un líder con esta información?

## 6. Extensión (Nivel intermedio – opcional)

Calcula medidas simples que apoyen la toma de decisiones.

```
# Promedio y variación de las ventas
promedio = datos['ventas'].mean()
maximo = datos['ventas'].max()
minimo = datos['ventas'].min()

promedio, maximo, minimo
```
