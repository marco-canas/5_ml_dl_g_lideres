### üìä Gr√°fico de barras: frecuencia de `INEQUALITY`



```python
import matplotlib.pyplot as plt

oecd_bli["INEQUALITY"].value_counts().plot(kind="bar")

plt.ylabel("Frecuencia")
plt.show()
```



üìå **Lectura del gr√°fico:**

### üìä Gr√°fico de barras: indicadores de bienestar


````python

```python
oecd_bli["INDICATOR"].value_counts().plot(kind="bar", figsize=(10,4))
plt.title("Indicadores del Better Life Index")
plt.xlabel("Indicador")
plt.ylabel("N√∫mero de registros")
plt.xticks(rotation=45)
plt.show()
```
````
