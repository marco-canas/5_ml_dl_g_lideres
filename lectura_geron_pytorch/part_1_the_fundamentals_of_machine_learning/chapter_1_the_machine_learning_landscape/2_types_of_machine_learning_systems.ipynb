# [Tipos de sistemas de aprendizaje automático](https://github.com/ageron/handson-mlp/blob/main/01_the_machine_learning_landscape.ipynb)


Existen tantos tipos diferentes de sistemas de aprendizaje automático que resulta útil clasificarlos en categorías generales, según los siguientes criterios:


* Cómo se guían durante el entrenamiento (supervisados, no supervisados, semisupervisados, autosupervisados ​​y otros).


* Si pueden aprender de forma incremental sobre la marcha (aprendizaje en línea o por lotes).


* Si funcionan simplemente comparando nuevos datos con datos conocidos o, en cambio, detectando patrones en los datos de entrenamiento y construyendo un modelo predictivo, como hacen los científicos (aprendizaje basado en instancias o basado en modelos).


Estos criterios no son excluyentes; se pueden combinar como se desee. 

Por ejemplo, un filtro de spam de última generación puede aprender sobre la marcha utilizando un modelo de red neuronal profunda entrenado con ejemplos de spam y ham proporcionados por personas; esto lo convierte en un sistema de aprendizaje supervisado en línea, basado en modelos.


Analicemos cada uno de estos criterios con más detalle. 

# Supervisión del Entrenamiento


Los sistemas de aprendizaje automático (ML) se pueden clasificar según la cantidad y el tipo de supervisión que reciben durante el entrenamiento. 

Existen muchas categorías, pero analizaremos las principales: 

* aprendizaje supervisado, 
* aprendizaje no supervisado, 
* aprendizaje autosupervisado, 
* aprendizaje semisupervisado y 
* aprendizaje por refuerzo.


# Aprendizaje supervisado


En el aprendizaje supervisado, el conjunto de entrenamiento que se introduce en el algoritmo incluye las soluciones deseadas, denominadas etiquetas (Figura 1-5).

Figure 1-5  

<img src = 'figure_1_5.png' width = 800> 

Una tarea típica de aprendizaje supervisado es la clasificación. 

El filtro de spam es un buen ejemplo de ello: se entrena con numerosos correos electrónicos de ejemplo junto con su clase (spam or ham) y debe aprender a clasificar los correos electrónicos nuevos.


Otra tarea típica es predecir un valor numérico objetivo, como el precio de un coche, dado un conjunto de características (kilometraje, antigüedad, marca, etc.). 

Este tipo de tarea se denomina regresión (Figura 1-6).¹


Figure 1-6

<img src = 'figure_1_6.png'>

Para entrenar el sistema, es necesario proporcionarle numerosos ejemplos de coches, incluyendo tanto sus características como sus objetivos (es decir, sus precios).


Tenga en cuenta que algunos modelos de regresión también pueden utilizarse para la clasificación, y viceversa. 

Por ejemplo, la regresión logística se utiliza habitualmente para la clasificación, ya que puede generar un valor que corresponde a la probabilidad de pertenecer a una clase determinada (p. ej., 20 % de probabilidad de ser spam).

Las palabras objetivo y etiqueta generalmente se consideran sinónimos en el aprendizaje supervisado, pero objetivo es más común en tareas de regresión y etiqueta en tareas de clasificación. 

Además, las características a veces se denominan predictores o atributos.

Estos términos pueden referirse a muestras individuales (p. ej., "la característica de kilometraje de este coche es igual a 15 000") o a todas las muestras (p. ej., "la característica de kilometraje está fuertemente correlacionada con el precio").



# Aprendizaje no supervisado


En el aprendizaje no supervisado, como se puede suponer, los datos de entrenamiento no están etiquetados. 

El sistema intenta aprender sin un profesor.



Por ejemplo, supongamos que tiene muchos datos sobre los visitantes de su blog. 

Quizás quiera ejecutar un algoritmo de agrupamiento para intentar detectar grupos de visitantes similares (Figura 1-7). 

<img src = 'figure_1_7.png'>

Las características pueden incluir la edad del usuario, su región, sus intereses, la duración de sus sesiones, etc. 

En ningún momento le dices al algoritmo a qué grupo pertenece un visitante: encuentra esas conexiones sin tu ayuda. 

Por ejemplo, podría notar que el 40% de tus visitantes son adolescentes amantes de los cómics y que suelen leer tu blog después de clase, mientras que el 20% son adultos que disfrutan de la ciencia ficción y lo visitan los fines de semana. 

Si utilizas un algoritmo de agrupamiento jerárquico, también podría subdividir cada grupo en grupos más pequeños. 

Esto puede ayudarte a orientar tus publicaciones para cada grupo.

Los algoritmos de visualización también son buenos ejemplos de aprendizaje no supervisado: se les proporciona una gran cantidad de datos complejos y sin etiquetar, y generan una representación 2D o 3D de los datos que se puede representar gráficamente fácilmente (Figura 1-8). 

<img src = 'figure_1_8.png' width = 700>

Estos algoritmos intentan preservar la estructura al máximo (por ejemplo, evitando que los clústeres separados en el espacio de entrada se superpongan en la visualización) para que se pueda comprender cómo se organizan los datos y, tal vez, identificar patrones inesperados.


Una tarea relacionada es la reducción de dimensionalidad, cuyo objetivo es simplificar los datos sin perder demasiada información. 

Una forma de lograrlo es fusionar varias características correlacionadas en una sola. 

Por ejemplo, el kilometraje de un coche puede estar fuertemente correlacionado con su antigüedad, por lo que el algoritmo de reducción de dimensionalidad las fusionará en una sola característica que representa el desgaste del coche. 

Esto se denomina extracción de características. 

Suele ser recomendable intentar reducir el número de dimensiones en los datos de entrenamiento mediante un algoritmo de reducción de dimensionalidad antes de introducirlos en otro algoritmo de aprendizaje automático (como un algoritmo de aprendizaje supervisado). 

Se ejecutará mucho más rápido, los datos ocuparán menos espacio en disco y memoria y, en algunos casos, incluso podrían tener un mejor rendimiento.

Otra tarea importante no supervisada es la detección de anomalías; por ejemplo, detectar transacciones inusuales con tarjetas de crédito para prevenir fraudes, detectar defectos de fabricación o eliminar automáticamente valores atípicos de un conjunto de datos antes de alimentarlo a otro algoritmo de aprendizaje. 

Durante el entrenamiento, el sistema muestra principalmente instancias normales, de modo que aprende a reconocerlas; luego, al detectar una nueva instancia, puede determinar si parece normal o si probablemente se trata de una anomalía (véase la Figura 1-9). 

Figure 1-9 
<img src = 'figure_1_9.png'>

Las características pueden incluir la distancia desde el hogar, la hora del día, el día de la semana, el importe retirado, la categoría del comercio, la frecuencia de las transacciones, etc. 

Una tarea muy similar es la detección de novedades: su objetivo es detectar nuevas instancias que se ven diferentes a todas las instancias del conjunto de entrenamiento. 

Esto requiere un conjunto de entrenamiento muy limpio, sin ninguna instancia que se desee que el algoritmo detecte. 

Por ejemplo, si tiene miles de fotos de perros, y el 1% de estas fotos representan chihuahuas, un algoritmo de detección de novedades no debería tratar las nuevas fotos de chihuahuas como novedades. 

Por otro lado, los algoritmos de detección de anomalías podrían considerar a estos perros tan raros y diferentes de otros perros que probablemente los clasificarían como anomalías (sin ofender a los chihuahuas).

Finalmente, otra tarea común no supervisada es el aprendizaje de reglas de asociación, cuyo objetivo es analizar grandes cantidades de datos y descubrir relaciones interesantes entre atributos. 

Por ejemplo, supongamos que usted es dueño de un supermercado. 

Ejecutar una regla de asociación en sus registros de ventas podría revelar que quienes compran salsa barbacoa y papas fritas también tienden a comprar bistec. 

Por lo tanto, podría ser conveniente colocar estos artículos cerca uno del otro.



# Aprendizaje semisupervisado


Dado que etiquetar datos suele ser lento y costoso, a menudo tendrá muchas instancias sin etiquetar y pocas instancias etiquetadas. 

Algunos algoritmos pueden procesar datos parcialmente etiquetados. 

Esto se denomina aprendizaje semisupervisado (Figura 1-10).

<img src = 'figure_1_10.png'>

Algunos servicios de alojamiento de fotos, como Google Photos, son buenos ejemplos de esto. 

Una vez que subes todas tus fotos familiares al servicio, este reconoce automáticamente que la misma persona A aparece en las fotos 1, 5 y 11, mientras que otra persona B aparece en las fotos 2, 5 y 7. 

Esta es la parte no supervisada del algoritmo (agrupamiento). 

Ahora, solo necesitas que el sistema le indique quiénes son estas personas. 

Simplemente agrega una etiqueta por persona 3 y podrá nombrar a todos los que aparecen en cada foto, lo cual es útil para buscar fotos.


La mayoría de los algoritmos de aprendizaje semisupervisado son combinaciones de algoritmos supervisados ​​y no supervisados. 

Por ejemplo, se puede usar un algoritmo de agrupamiento para agrupar instancias similares, y luego cada instancia sin etiquetar se puede etiquetar con la etiqueta más común en su grupo. 

Una vez etiquetado todo el conjunto de datos, es posible usar cualquier algoritmo de aprendizaje supervisado. 


# Aprendizaje autosupervisado


Otro enfoque del aprendizaje automático implica generar un conjunto de datos completamente etiquetado a partir de uno completamente sin etiquetar. 

Una vez etiquetado todo el conjunto de datos, se puede utilizar cualquier algoritmo de aprendizaje supervisado. 

Este enfoque se denomina aprendizaje autosupervisado.


Por ejemplo, si se tiene un gran conjunto de datos de imágenes sin etiquetar, se puede enmascarar aleatoriamente una pequeña parte de cada imagen y luego entrenar un modelo para recuperar la imagen original (Figura 1-11). 

<img src = 'figure_1_11.png'>

Durante el entrenamiento, las imágenes enmascaradas se utilizan como entrada del modelo y las imágenes originales como etiquetas.

El modelo resultante puede ser bastante útil en sí mismo, por ejemplo, para reparar imágenes dañadas o borrar objetos no deseados. 

Pero, en la mayoría de los casos, un modelo entrenado mediante aprendizaje autosupervisado no es el objetivo final. 

Normalmente, querrá ajustar y perfeccionar el modelo para una tarea ligeramente diferente, una que realmente le interese.



Por ejemplo, supongamos que lo que realmente desea es tener un modelo de clasificación de mascotas: dada la foto de cualquier mascota, le indicará a qué especie pertenece. 

Si tiene un gran conjunto de datos de fotos de mascotas sin etiquetar, puede comenzar entrenando un modelo de reparación de imágenes mediante aprendizaje autosupervisado. 

Una vez que funcione bien, debería ser capaz de distinguir diferentes especies de mascotas: cuando repara la imagen de un gato con la cara enmascarada, debe saber que no debe agregar la cara de un perro. 

Suponiendo que la arquitectura de su modelo lo permita (y la mayoría de las arquitecturas de redes neuronales lo permiten), es posible ajustar el modelo para que prediga las especies de mascotas en lugar de reparar imágenes. 

El paso final consiste en ajustar el modelo con un conjunto de datos etiquetado: el modelo ya sabe cómo son los gatos, perros y otras especies de mascotas, por lo que este paso solo es necesario para que el modelo pueda aprender la correspondencia entre las especies que ya conoce y las etiquetas que esperamos.


Transferir conocimiento de una tarea a otra se denomina aprendizaje por transferencia y es una de las técnicas más importantes del aprendizaje automático actual, especialmente cuando se utilizan redes neuronales profundas (es decir, redes neuronales compuestas por muchas capas de neuronas). 

Analizaremos esto en detalle en la Parte II.


Como veremos en el Capítulo 15, los modelos de lenguaje grandes (LLM) se entrenan de forma muy similar, enmascarando palabras aleatorias en un corpus de texto enorme y entrenando el modelo para predecir las palabras que faltan. 

Este gran modelo preentrenado puede ajustarse para diversas aplicaciones, desde el análisis de sentimientos hasta los chatbots.


Algunas personas consideran que el aprendizaje autosupervisado forma parte del aprendizaje no supervisado, ya que trabaja con conjuntos de datos sin etiquetas. 

Sin embargo, el aprendizaje autosupervisado utiliza etiquetas (generadas) durante el entrenamiento, por lo que, en ese sentido, se acerca más al aprendizaje supervisado. 

El término "aprendizaje no supervisado" se utiliza generalmente para tareas como la agrupación en clústeres, la reducción de dimensionalidad o la detección de anomalías, mientras que el aprendizaje autosupervisado se centra en las mismas tareas que el aprendizaje supervisado: principalmente la clasificación y la regresión. 

En resumen, es mejor tratar el aprendizaje autosupervisado como una categoría independiente.


# Aprendizaje por refuerzo


El aprendizaje por refuerzo es un campo muy diferente. 

El sistema de aprendizaje, llamado agente en este contexto, puede observar el entorno, seleccionar y ejecutar acciones, y obtener recompensas a cambio (o penalizaciones en forma de recompensas negativas, como se muestra en la Figura 1-12). 

<img src = 'figure_1_12.png'>

Luego, debe aprender por sí mismo cuál es la mejor estrategia, llamada política, para obtener la mayor recompensa a lo largo del tiempo. 

Una política define qué acción debe elegir el agente en una situación determinada.

Por ejemplo, muchos robots implementan algoritmos de aprendizaje por refuerzo para aprender a caminar. 

El programa AlphaGo de DeepMind también es un buen ejemplo de aprendizaje por refuerzo: fue noticia en mayo de 2017 cuando venció a Ke Jie, el jugador número uno del mundo en ese momento, en la partida de Go. 

Aprendió su estrategia ganadora analizando millones de partidas y luego jugando muchas contra sí mismo. 

Cabe destacar que el aprendizaje se desactivó durante las partidas contra el campeón; AlphaGo simplemente aplicaba la estrategia aprendida. 

Como se verá en la siguiente sección, esto se denomina aprendizaje fuera de línea.



# Aprendizaje por lotes versus aprendizaje en línea



Otro criterio utilizado para clasificar los sistemas de aprendizaje automático es si el sistema puede aprender de forma incremental a partir de un flujo de datos entrante. 

Por ejemplo, los bosques aleatorios (véase el Capítulo 6) solo se pueden entrenar desde cero con el conjunto de datos completo (esto se denomina aprendizaje por lotes), mientras que otros modelos se pueden entrenar con un lote de datos a la vez, por ejemplo, mediante el descenso de gradiente (véase el Capítulo 4); esto se denomina aprendizaje en línea. 

# Aprendizaje por lotes


En el aprendizaje por lotes, el sistema debe entrenarse con todos los datos disponibles. 

Esto generalmente requiere mucho tiempo y recursos informáticos, por lo que suele realizarse sin conexión.


Primero se entrena el sistema y luego se lanza a producción y funciona sin aprender más; simplemente aplica lo aprendido. 

Esto se denomina aprendizaje sin conexión.


Desafortunadamente, el rendimiento de un modelo tiende a deteriorarse lentamente con el tiempo, simplemente porque el mundo continúa evolucionando mientras que el modelo permanece inalterado. 

Este fenómeno a menudo se denomina deriva de datos (o deterioro del modelo).

La solución es reentrenar regularmente el modelo con datos actualizados. 

La frecuencia con la que se debe hacer esto depende del caso de uso: si el modelo clasifica imágenes de gatos y perros, su rendimiento se deteriorará muy lentamente, pero si el modelo trabaja con sistemas que evolucionan rápidamente, por ejemplo, al realizar predicciones sobre el mercado financiero, es probable que se deteriore con bastante rapidez. 

Incluso un modelo entrenado para clasificar imágenes de gatos y perros puede necesitar ser reentrenado regularmente, no porque los gatos y los perros muten de la noche a la mañana, sino porque las cámaras cambian constantemente, junto con los formatos de imagen, la nitidez, el brillo y las proporciones de tamaño. 

Además, puede que el próximo año la gente adore razas diferentes, o que decida vestir a sus mascotas con sombreros diminutos, ¿quién sabe?


Si desea que un sistema de aprendizaje por lotes detecte nuevos datos (como un nuevo tipo de spam), necesita entrenar una nueva versión del sistema desde cero con el conjunto de datos completo (no solo los nuevos, sino también los antiguos) y luego reemplazar el modelo antiguo por el nuevo. 

Afortunadamente, todo el proceso de entrenamiento, evaluación y lanzamiento de un sistema de aprendizaje automático se puede automatizar (como vimos en la Figura 1-3), por lo que incluso un sistema de aprendizaje por lotes puede adaptarse al cambio. 

Simplemente actualice los datos y entrene una nueva versión del sistema desde cero tantas veces como sea necesario. 

Esta solución es sencilla y suele funcionar bien, pero entrenar con el conjunto completo de datos puede llevar muchas horas, por lo que normalmente solo se entrenaría un nuevo sistema cada 24 horas o incluso semanalmente. 

Si su sistema necesita adaptarse a datos que cambian rápidamente (por ejemplo, para predecir los precios de las acciones), necesita una solución más reactiva. 

Además, entrenar con el conjunto completo de datos requiere muchos recursos informáticos (CPU, espacio de memoria, espacio de disco, E/S de disco, E/S de red, etc.). 

Si tiene muchos datos y automatiza su sistema para entrenar desde cero a diario, acabará costando mucho dinero. 

Si la cantidad de datos es enorme y su sistema debe estar siempre actualizado, puede incluso resultar imposible utilizar el aprendizaje por lotes.

Finalmente, si su sistema necesita aprender de forma autónoma y cuenta con recursos limitados (por ejemplo, una aplicación para smartphones o un rover en Marte), entonces transportar grandes cantidades de datos de entrenamiento y consumir muchos recursos para entrenar durante horas todos los días es un problema.


Una mejor opción en todos estos escenarios es utilizar algoritmos capaces de aprender de forma incremental.


# Aprendizaje en línea


En el aprendizaje en línea, el sistema se entrena de forma incremental alimentándolo con instancias de datos secuencialmente, ya sea individualmente o en pequeños grupos llamados minilotes. 

Cada paso de aprendizaje es rápido y económico, por lo que el sistema puede aprender sobre los nuevos datos sobre la marcha, a medida que llegan (véase la Figura 1-13). 

<img src = 'figure_1_13.png' width = 800>

El algoritmo en línea más común, con diferencia, es el descenso de gradiente, pero existen algunos otros.

El aprendizaje en línea es útil para sistemas que necesitan adaptarse a cambios con extrema rapidez (por ejemplo, para detectar nuevos patrones en el mercado de valores). 

También es una buena opción si se dispone de recursos informáticos limitados; por ejemplo, si el modelo se entrena en un dispositivo móvil.


Y lo que es más importante, los algoritmos de aprendizaje en línea pueden utilizarse para entrenar modelos con grandes conjuntos de datos que no caben en la memoria de una sola máquina (esto se denomina aprendizaje fuera del núcleo).


El algoritmo carga parte de los datos, ejecuta un paso de entrenamiento con ellos y repite el proceso hasta que se haya ejecutado con todos los datos (véase la Figura 1-14).

<img src = 'figure_1_14.png'>

Un parámetro importante de los sistemas de aprendizaje en línea es la rapidez con la que deben adaptarse a los datos cambiantes: esto se denomina tasa de aprendizaje. 

Si se establece una tasa de aprendizaje alta, el sistema se adaptará rápidamente a los nuevos datos, pero también tenderá a olvidar rápidamente los antiguos: esto se denomina olvido catastrófico (o interferencia catastrófica). 

No conviene que un filtro de spam marque solo los tipos de spam más recientes que se le hayan mostrado. 

Por el contrario, si se establece una tasa de aprendizaje baja, el sistema tendrá mayor inercia; es decir, aprenderá más lentamente, pero también será menos sensible al ruido en los nuevos datos o a secuencias de puntos de datos no representativos (valores atípicos).


El aprendizaje fuera del núcleo suele realizarse sin conexión (es decir, no en el sistema en vivo), por lo que el aprendizaje en línea puede ser un nombre confuso. 

Considérelo como aprendizaje incremental. 

Además, los minilotes a menudo se denominan simplemente "lotes", por lo que el aprendizaje por lotes también es un nombre confuso. 

Considérelo como aprender desde cero con el conjunto de datos completo. 

Un gran desafío del aprendizaje en línea es que, si se introducen datos incorrectos en el sistema, su rendimiento disminuirá, posiblemente rápidamente (dependiendo de la calidad de los datos y la velocidad de aprendizaje). 

Si se trata de un sistema en vivo, sus clientes lo notarán. 

Por ejemplo, los datos incorrectos podrían provenir de un error (p. ej., un sensor defectuoso en un robot) o de alguien que intenta manipular el sistema (p. ej., inundar un motor de búsqueda para intentar posicionarse bien en los resultados). 

Para reducir este riesgo, es necesario supervisar el sistema de cerca y desactivar el aprendizaje rápidamente (y posiblemente volver a un estado operativo anterior) si se detecta una disminución en el rendimiento. 

También es recomendable supervisar los datos de entrada y reaccionar ante datos anormales; por ejemplo, utilizando un algoritmo de detección de anomalías (véase el Capítulo 8).


# Aprendizaje basado en instancias frente a aprendizaje basado en modelos  

Otra forma de clasificar los sistemas de aprendizaje automático es según su generalización.


La mayoría de las tareas de aprendizaje automático consisten en realizar predicciones. 

Esto significa que, dados varios ejemplos de entrenamiento, el sistema debe ser capaz de realizar buenas predicciones (generalizar a) ejemplos que nunca ha visto. 

Contar con una buena medición de rendimiento en los datos de entrenamiento es bueno, pero insuficiente; el verdadero objetivo es obtener un buen rendimiento en nuevas instancias.


Existen dos enfoques principales para la generalización: el aprendizaje basado en instancias y el aprendizaje basado en modelos.


# Aprendizaje basado en instancias


Posiblemente, la forma más sencilla de aprendizaje sea simplemente memorizar. 

Si se creara un filtro de spam de esta manera, este simplemente marcaría todos los correos electrónicos idénticos a los que ya han sido marcados por los usuarios; no es la peor solución, pero ciertamente no la mejor.


En lugar de marcar solo los correos electrónicos idénticos a los correos electrónicos de spam conocidos, el filtro de spam podría programarse para que también marque los correos electrónicos muy similares a los correos electrónicos de spam conocidos. 

Esto requiere una medida de similitud entre dos correos electrónicos. 

Una medida de similitud (muy básica) entre dos correos electrónicos podría ser contar el número de palabras que tienen en común. 

El sistema marcaría un correo electrónico como spam si tiene muchas palabras en común con un correo electrónico spam conocido.


Esto se denomina aprendizaje basado en instancias: el sistema memoriza los ejemplos y luego los generaliza a nuevos casos utilizando una medida de similitud para compararlos con los ejemplos aprendidos (o un subconjunto de ellos). 

Por ejemplo, en la Figura 1-15, la nueva instancia se clasificaría como un triángulo porque la mayoría de las instancias más similares pertenecen a esa clase.


<img src = 'figure_1_15.png' width = 800>

El aprendizaje basado en instancias suele ser eficaz con conjuntos de datos pequeños, especialmente si los datos cambian constantemente, pero no es muy escalable: requiere implementar una copia completa del conjunto de entrenamiento en producción; realizar predicciones requiere buscar instancias similares, lo que puede ser bastante lento; y no funciona bien con datos de alta dimensión, como imágenes.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
lifesat 
```

```python
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values

# Visualize the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()

# Select a linear model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction for Puerto Rico 
X_new = [[33_442.8]]  # Puerto Rico' GDP per capita in 2020
print(model.predict(X_new)) # outputs [[6.01610329]]
```

¡Parece haber una tendencia! Aunque los datos son ruidosos (es decir, parcialmente aleatorios), parece que la satisfacción con la vida aumenta de forma más o menos lineal a medida que aumenta el PIB per cápita del país. 

Por lo tanto, decide modelar la satisfacción con la vida como una función lineal del PIB per cápita (asume que cualquier desviación de esa línea es simplemente ruido aleatorio). 

Este paso se denomina selección del modelo: se seleccionó un modelo lineal de satisfacción con la vida con un solo atributo, el PIB per cápita (Ecuación 1-1).

Equation 1-1. A simple linear model
$$
\text{life satisfaction}
=
\theta_{0}
+
\theta_{1}\,\text{GDP per capita}
$$


$$
\mathrm{life\_satisfaction}
=
\theta_{0}
+
\theta_{1}\,\mathrm{GDP\_per\_capita}
$$


Este modelo tiene dos parámetros, $\theta_{0}$ y $\theta_{1}$.

Al ajustar estos parámetros, puede lograr que su modelo represente cualquier función lineal, como se muestra en la Figura 1-18.

<img src = 'figure_1_18.png'>

Antes de poder usar su modelo, necesita definir los valores de los parámetros $\theta_{0}$ (intersecto en el origen o término de sesgo) y $\theta_{1}$ (pendiente de la recta).



¿Cómo puede saber qué valores harán que su modelo tenga el mejor rendimiento? 

Para responder a esta pregunta, necesita especificar una medida de rendimiento. 

Puede definir una función de utilidad (o función de aptitud) que mida la calidad de su modelo, o una función de costo (también conocida como función de pérdida) que mida su calidad. 

Para problemas de regresión lineal, se suele usar una función de costo que mide la distancia entre las predicciones del modelo lineal y los ejemplos de entrenamiento; el objetivo es minimizar esta distancia.



Aquí es donde entra en juego el algoritmo de regresión lineal: se le introducen los ejemplos de entrenamiento y este encuentra los parámetros que hacen que el modelo lineal se ajuste mejor a los datos.



Esto se denomina entrenamiento del modelo. 

En nuestro caso, el algoritmo encuentra que los valores óptimos de los parámetros son $\theta_{0} = 3,75$ y $\theta_{1} = 6,78 × 10–5$.


De forma confusa, la palabra "modelo" puede referirse a un tipo de modelo (p. ej., regresión lineal), a una arquitectura de modelo completamente especificada (p. ej., regresión lineal con una entrada y una salida) o al modelo final entrenado, listo para usarse en predicciones (p. ej., regresión lineal con una entrada y una salida, utilizando $\theta_{0} = 3,75$ y $\theta_{1} = 6,78 × 10–5$).


La ​​selección del modelo consiste en elegir el tipo de modelo y especificar completamente su arquitectura. 

Entrenar un modelo significa ejecutar un algoritmo para encontrar los parámetros del modelo que mejor se ajusten a los datos de entrenamiento y, con suerte, realizar buenas predicciones con los nuevos datos.


Ahora, el modelo se ajusta a los datos de entrenamiento lo más fielmente posible (para un modelo lineal), como se puede ver en la Figura 1-19.

# Cantidad insuficiente de datos de entrenamiento


Para que un niño pequeño aprenda qué es una manzana, basta con señalarla y decir "manzana" (quizás repitiendo este procedimiento varias veces).

Incluso para problemas muy simples, se suelen necesitar miles de ejemplos, y para problemas complejos como el reconocimiento de imágenes o voz, pueden necesitarse millones de ejemplos (a menos que se puedan reutilizar partes de un modelo existente, es decir, aprendizaje por transferencia).


# La ​​irrazonable eficacia de los datos


En un famoso artículo publicado en 2001, los investigadores de Microsoft Michele Banko y Eric Brill demostraron que algoritmos de aprendizaje automático muy diferentes, incluidos algunos bastante simples, funcionaban casi de manera idéntica en un problema complejo de desambiguación del lenguaje natural una vez que se les proporcionaban suficientes datos (como se puede ver en la Figura 1-21).

<img src = 'figure_1_21.png' width = 600>

Como lo expresan los autores, «estos resultados sugieren que deberíamos reconsiderar la compensación entre invertir tiempo y dinero en el desarrollo de algoritmos y en el desarrollo de corpus».


La idea de que los datos son más importantes que los algoritmos para problemas complejos fue popularizada por Peter Norvig et al. en un artículo titulado «La irrazonable efectividad de los datos», publicado en 2009.

Cabe señalar, sin embargo, que los conjuntos de datos pequeños y medianos siguen siendo muy comunes, y no siempre es fácil ni económico obtener datos de entrenamiento adicionales, así que no abandone los algoritmos todavía.

# Datos de entrenamiento no representativos


Para generalizar correctamente, es crucial que los datos de entrenamiento sean representativos de los nuevos casos a los que se desea generalizar. 

Esto se aplica tanto al aprendizaje basado en instancias como al aprendizaje basado en modelos.


Por ejemplo, el conjunto de países que se utilizó anteriormente para entrenar el modelo lineal no era perfectamente representativo; no contenía ningún país con un PIB per cápita inferior a $23.500$ ni superior a $62.500$. 

La Figura 1-22 muestra el aspecto de los datos al añadir dichos países.


<img src = 'figure_1_22.png'>

Si se entrena un modelo lineal con estos datos, se obtiene la línea continua, mientras que el modelo anterior se representa con la línea de puntos. 

Como se puede observar, añadir algunos países no incluidos no solo altera significativamente el modelo, sino que también deja claro que un modelo lineal tan simple probablemente nunca funcionará bien. 

Parece que los países muy ricos no son más felices que los países moderadamente ricos (de hecho, ¡parecen ligeramente más infelices!) y, por el contrario, algunos países pobres parecen más felices que muchos países ricos.

Al utilizar un conjunto de entrenamiento no representativo, se entrenó un modelo que probablemente no genere predicciones precisas, especialmente para países muy pobres y muy ricos.


Es crucial utilizar un conjunto de entrenamiento que sea representativo de los casos a los que se desea generalizar.

Esto suele ser más difícil de lo que parece: si la muestra es demasiado pequeña, se producirá ruido de muestreo (es decir, datos no representativos debido al azar), pero incluso muestras muy grandes pueden ser no representativas si el método de muestreo es defectuoso. 

Esto se denomina **sesgo de muestreo.**

# Ejemplos de sesgo de muestreo


Quizás el ejemplo más famoso de sesgo de muestreo ocurrió durante las elecciones presidenciales estadounidenses de 1936, que enfrentaron a Landon contra Roosevelt: el Literary Digest realizó una encuesta muy amplia, enviando correos a unos 10 millones de personas. 

Obtuvo 2,4 millones de respuestas y predijo con gran certeza que Landon obtendría el 57 % de los votos.


En cambio, Roosevelt ganó con el 62 % de los votos. 

La falla residió en el método de muestreo del Literary Digest:



* En primer lugar, para obtener las direcciones a las que enviar las encuestas, el Literary Digest utilizó directorios telefónicos, listas de suscriptores de revistas, listas de miembros de clubes, etc. 

Todas estas listas tendían a favorecer a las personas más adineradas, quienes eran más propensas a votar por los republicanos (de ahí el voto de Landon).



* En segundo lugar, menos del 25 % de los encuestados respondió. 

Nuevamente, esto introdujo un sesgo de muestreo, al descartar potencialmente a personas a las que no les interesaba mucho la política, a quienes no les gustaba el Literary Digest y a otros grupos clave. 

Este es un tipo especial de sesgo de muestreo llamado sesgo de falta de respuesta.


Aquí hay otro ejemplo: supongamos que quieres crear un sistema para reconocer videos de música funk. 

Una forma de crear tu conjunto de entrenamiento es buscar "música funk" en YouTube y usar los videos resultantes. 

Pero esto supone que el motor de búsqueda de YouTube devuelve un conjunto de videos representativo de todos los videos de música funk disponibles en YouTube. 

En realidad, es probable que los resultados de la búsqueda estén sesgados hacia artistas populares (y si vives en Brasil, encontrarás muchos videos de "funk carioca", que no suenan como James Brown). 

Por otro lado, ¿de qué otra manera puedes obtener un conjunto de entrenamiento amplio?

# Datos de baja calidad


Obviamente, si sus datos de entrenamiento están llenos de errores, valores atípicos y ruido (por ejemplo, debido a mediciones de baja calidad), será más difícil para el sistema detectar los patrones subyacentes, por lo que es menos probable que funcione bien. 

A menudo, vale la pena dedicar tiempo a depurar los datos de entrenamiento. 

Lo cierto es que la mayoría de los científicos de datos dedican gran parte de su tiempo a ello. 

A continuación, se muestran algunos ejemplos de cuándo conviene depurar los datos de entrenamiento:



* Si algunas instancias son claramente valores atípicos, puede ser útil simplemente descartarlas o intentar corregir los errores manualmente.



* Si a algunas instancias les faltan algunas características (por ejemplo, el 5 % de sus clientes no especificó su edad), debe decidir si desea ignorar este atributo por completo, ignorar estas instancias, completar los valores faltantes (por ejemplo, con la mediana de edad) o entrenar un modelo con la característica y otro sin ella. 

# Características Irrelevantes


Como dice el refrán: basura entra, basura sale. 

Su sistema solo podrá aprender si los datos de entrenamiento contienen suficientes características relevantes y no demasiadas irrelevantes. 

Una parte fundamental del éxito de un proyecto de aprendizaje automático es generar un buen conjunto de características para entrenar. 

Este proceso, llamado ingeniería de características, implica los siguientes pasos:


* Selección de características (seleccionar las características más útiles para entrenar entre las existentes)



* Extracción de características (combinar características existentes para producir una más útil; como vimos anteriormente, los algoritmos de reducción de dimensionalidad pueden ayudar)



• Creación de nuevas características mediante la recopilación de nuevos datos.

## 1. ¿Qué significa *GDP per capita*?

**GDP per capita** significa **Producto Interno Bruto per cápita**.

Formalmente:

$$
\text{GDP per capita} = \frac{\text{PIB total del país}}{\text{Número de habitantes}}
$$

* Es una **medida promedio de producción/ingreso por persona**.
* Se usa como **proxy del nivel de vida** de un país.
* Permite **comparar países** con poblaciones muy distintas.



## Importante (y muy didáctico):

* No garantiza bienestar individual
* Pero es muy útil como **variable explicativa** en modelos sencillos

```python
from pathlib import Path
import urllib.request

datapath = Path() / "datasets" / "lifesat"
datapath.mkdir(parents=True, exist_ok=True) # crea una carpeta con 4 dataset sobre safisfacción con la vida

data_root = "https://github.com/ageron/data/raw/main/"
for filename in ("oecd_bli.csv", "gdp_per_capita.csv"):
    if not (datapath / filename).is_file():
        print("Downloading", filename)
        url = data_root + "lifesat/" + filename
        urllib.request.urlretrieve(url, datapath / filename)
```

```python
import pandas as pd 
oecd_bli = pd.read_csv(datapath / "oecd_bli.csv")
gdp_per_capita = pd.read_csv(datapath / "gdp_per_capita.csv")
```

```python
oecd_bli
```

```python
gdp_per_capita
```

```python
gdp_year = 2020
gdppc_col = "GDP per capita (USD)"
lifesat_col = "Life satisfaction"

gdp_per_capita = gdp_per_capita[gdp_per_capita["Year"] == gdp_year]
gdp_per_capita = gdp_per_capita.drop(["Code", "Year"], axis=1)
gdp_per_capita.columns = ["Country", gdppc_col]
gdp_per_capita.set_index("Country", inplace=True)

gdp_per_capita.head()
```

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")

oecd_bli.head()
```

```python
full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                              left_index=True, right_index=True)
full_country_stats.sort_values(by=gdppc_col, inplace=True)
full_country_stats = full_country_stats[[gdppc_col, lifesat_col]]

full_country_stats.head()
```

Los modelos complejos, como las redes neuronales profundas, pueden detectar patrones sutiles en los datos. 

Sin embargo, si el conjunto de entrenamiento presenta ruido o es demasiado pequeño, lo que introduce ruido de muestreo, es probable que el modelo detecte patrones en el propio ruido (como en el ejemplo del taxista).

Obviamente, estos patrones no se generalizarán a nuevas instancias. 

Por ejemplo, supongamos que introduce en su modelo de satisfacción vital muchos más atributos, incluidos los poco informativos, como el nombre del país. 

En ese caso, un modelo complejo podría detectar patrones como el hecho de que todos los países de los datos de entrenamiento con una "w" en su nombre tengan una satisfacción vital superior a 7: Nueva Zelanda (7,3), Noruega (7,6), Suecia (7,3) y Suiza (7,5). 

¿Qué tan seguro está de que la regla de satisfacción con la "w" se generalice a Ruanda o Zimbabue? 

Obviamente, este patrón se produjo en los datos de entrenamiento por pura casualidad, pero el modelo no tiene forma de determinar si un patrón es real o simplemente el resultado del ruido en los datos. 

El sobreajuste se produce cuando el modelo es demasiado complejo en relación con la cantidad y el ruido de los datos de entrenamiento, por lo que empieza a aprender patrones aleatorios en ellos. 

Posibles soluciones:



* Simplificar el modelo seleccionando uno con menos parámetros (p. ej., un modelo lineal en lugar de un modelo polinomial de alto grado), reduciendo el número de atributos en los datos de entrenamiento o restringiéndolo.



• Recopilar más datos de entrenamiento.



* Reducir el ruido en los datos de entrenamiento (p. ej., corregir errores de datos y eliminar valores atípicos).

Restringir un modelo para simplificarlo y reducir el riesgo de sobreajuste se denomina regularización. 

Por ejemplo, el modelo lineal que definimos anteriormente tiene dos parámetros, θ0 y θ1. 

Esto otorga al algoritmo de aprendizaje dos grados de libertad para adaptar el modelo a los datos de entrenamiento: puede ajustar tanto la altura (θ0) como la pendiente (θ1) de la línea. 

Si forzáramos θ1 = 0, el algoritmo solo tendría un grado de libertad y tendría muchas más dificultades para ajustar los datos correctamente: solo podría mover la línea hacia arriba o hacia abajo para acercarse lo más posible a las instancias de entrenamiento, de modo que terminaría alrededor de la media. 

¡Un modelo muy simple, sin duda! Si permitimos que el algoritmo modifique θ1, pero lo forzamos a mantenerlo pequeño, el algoritmo de aprendizaje tendrá entre uno y dos grados de libertad. 

Producirá un modelo más simple que uno con dos grados de libertad, pero más complejo que uno con solo uno. 

Se busca el equilibrio adecuado entre un ajuste perfecto de los datos de entrenamiento y una simplicidad del modelo suficiente para garantizar una buena generalización.


La Figura 1-24 muestra tres modelos. La línea punteada representa el modelo original, entrenado con los países representados como círculos (sin los países representados como cuadrados); la línea continua es nuestro segundo modelo, entrenado con todos los países (círculos y cuadrados); y la línea discontinua es un modelo entrenado con los mismos datos que el primero, pero con una restricción de regularización. 

Se puede observar que la regularización obligó al modelo a tener una pendiente menor: este modelo no se ajusta a los datos de entrenamiento (círculos) tan bien como el primero, pero se generaliza mejor a nuevos ejemplos que no detectó durante el entrenamiento (cuadrados).

La cantidad de regularización que se aplica durante el aprendizaje se puede controlar mediante un hiperparámetro. 

Un hiperparámetro es un parámetro de un algoritmo de aprendizaje (no del modelo). 

Por lo tanto, no se ve afectado por el algoritmo de aprendizaje en sí; debe configurarse antes del entrenamiento y permanece constante durante este. 

Si se establece el hiperparámetro de regularización en un valor muy alto, se obtendrá un modelo casi plano (una pendiente cercana a cero); es casi seguro que el algoritmo de aprendizaje no sobreajustará los datos de entrenamiento, pero será menos probable que encuentre una buena solución. 

Ajustar los hiperparámetros es una parte importante del desarrollo de un sistema de aprendizaje automático (verá un ejemplo detallado en el siguiente capítulo).


# Subajuste de los datos de entrenamiento


Como puede imaginar, el subajuste es lo opuesto al sobreajuste: ocurre cuando el modelo es demasiado simple para comprender la estructura subyacente de los datos. 

Por ejemplo, un modelo lineal de satisfacción con la vida es propenso al subajuste; la realidad es simplemente más compleja que el modelo, por lo que sus predicciones son inexactas, incluso en los ejemplos de entrenamiento. Estas son las principales opciones para solucionar este problema:



• Seleccionar un modelo más potente, con más parámetros.



• Incorporar mejores características al algoritmo de aprendizaje (ingeniería de características).



• Reducir las restricciones del modelo (por ejemplo, reduciendo el hiperparámetro de regularización).


# Problemas de implementación


Incluso si se cuenta con un conjunto de datos grande y limpio, y se logra entrenar un modelo atractivo que no se sobreajusta ni se infraajusta a los datos, es posible que surjan problemas durante la implementación: por ejemplo, el modelo puede ser demasiado complejo de mantener, demasiado grande para caber en la memoria, demasiado lento, no escalar correctamente, presentar vulnerabilidades de seguridad, quedar obsoleto si no se actualiza con la frecuencia suficiente, etc.


En resumen, un proyecto de aprendizaje automático implica mucho más que datos y modelos. 

Sin embargo, las habilidades necesarias para gestionar estos problemas operativos son bastante diferentes a las requeridas para el modelado de datos, por lo que las empresas suelen contar con un equipo de MLOps (operaciones de ML) dedicado a ello.

Hay un último tema importante que abordar: una vez que haya entrenado un modelo, no debe simplemente "esperar" que se generalice a nuevos casos. Quiere evaluarlo y ajustarlo si es necesario. Veamos cómo hacerlo.


# Pruebas y validación


La única manera de saber qué tan bien se generalizará un modelo a nuevos casos es probarlo en ellos. 

Una forma de hacerlo es poner el modelo en producción y monitorear su rendimiento. 

Esto funciona bien, pero si su modelo es extremadamente deficiente, sus usuarios se quejarán, lo cual no es la mejor idea.


Una mejor opción es dividir los datos en dos conjuntos: el conjunto de entrenamiento y el conjunto de prueba.


Como sus nombres lo indican, se entrena el modelo con el conjunto de entrenamiento y se prueba con el conjunto de prueba. 

La tasa de error en nuevos casos se denomina error de generalización (o error fuera de muestra), y al evaluar el modelo con el conjunto de prueba, se obtiene una estimación de este error. 

Este valor indica el rendimiento del modelo en instancias que nunca antes ha visto. 

Si el error de entrenamiento es bajo (es decir, su modelo comete pocos errores en el conjunto de entrenamiento) pero el error de generalización es alto, significa que su modelo está sobreajustando los datos de entrenamiento.


Es común usar el 80% de los datos para el entrenamiento y reservar el 20% para las pruebas. 

Sin embargo, esto depende del tamaño del conjunto de datos: si contiene 10 millones de instancias, reservar el 1% significa que su conjunto de pruebas contendrá 100,000 instancias, probablemente más que suficiente para obtener una buena estimación del error de generalización.


# Ajuste de hiperparámetros y selección del modelo


Evaluar un modelo es bastante sencillo: simplemente use un conjunto de pruebas. B

Ahora supongamos que el modelo lineal se generaliza mejor, pero se desea aplicar regularización para evitar el sobreajuste. 

La pregunta es: ¿cómo se elige el valor del hiperparámetro de regularización? Una opción es entrenar 100 modelos diferentes utilizando 100 valores distintos para este hiperparámetro. 

Supongamos que se encuentra el mejor valor de hiperparámetro que produce un modelo con el menor error de generalización (digamos, solo un 5%). 

Se lanza este modelo a producción, pero lamentablemente no funciona tan bien como se esperaba y produce un 15% de errores. ¿Qué sucedió?


El problema es que se midió el error de generalización varias veces en el conjunto de prueba y se adaptaron el modelo y los hiperparámetros para producir el mejor modelo para ese conjunto en particular. Esto significa que es poco probable que el modelo funcione tan bien con datos nuevos.


Una solución común a este problema se denomina validación de retención (Figura 1-25): simplemente se retiene parte del conjunto de entrenamiento para evaluar varios modelos candidatos y seleccionar el mejor. 

El nuevo conjunto de reserva se denomina conjunto de validación (o conjunto de desarrollo o conjunto dev). 

Más específicamente, se entrenan múltiples modelos con diversos hiperparámetros en el conjunto de entrenamiento reducido (es decir, el conjunto de entrenamiento completo menos el conjunto de validación) y se selecciona el modelo con mejor rendimiento en el conjunto de validación.


Tras este proceso de validación de reserva, se entrena el mejor modelo en el conjunto de entrenamiento completo (incluido el conjunto de validación), lo que da como resultado el modelo final. 

Por último, se evalúa este modelo final en el conjunto de prueba para obtener una estimación del error de generalización.

Esta solución suele funcionar bastante bien. 

Sin embargo, si el conjunto de validación es demasiado pequeño, las evaluaciones del modelo serán imprecisas: podría terminar seleccionando un modelo subóptimo por error. 

Por el contrario, si el conjunto de validación es demasiado grande, el conjunto de entrenamiento restante será mucho menor que el conjunto de entrenamiento completo. 

¿Por qué es esto malo? Bueno, dado que el modelo final se entrenará con el conjunto de entrenamiento completo, no es ideal comparar modelos candidatos entrenados con un conjunto de entrenamiento mucho más pequeño. 

Sería como seleccionar al velocista más rápido para participar en una maratón. 

Una forma de resolver este problema es realizar una validación cruzada repetida, utilizando muchos conjuntos de validación pequeños. 

Cada modelo se evalúa una vez por conjunto de validación después de entrenarse con el resto de los datos. 

Al promediar todas las evaluaciones de un modelo, se obtiene una medida mucho más precisa de su rendimiento. 

Sin embargo, existe una desventaja: el tiempo de entrenamiento se multiplica por el número de conjuntos de validación. 

# Discordancia de datos


En algunos casos, es fácil obtener una gran cantidad de datos para el entrenamiento, pero estos datos probablemente no serán perfectamente representativos de los datos que se utilizarán en producción.


Por ejemplo, supongamos que desea crear una aplicación móvil para tomar fotos de flores y determinar automáticamente su especie. 

Puede descargar fácilmente millones de fotos de flores de la web, pero no serán perfectamente representativas de las fotos que se tomarán realmente con la aplicación en un dispositivo móvil. 

Quizás solo tenga 1000 fotos representativas (es decir, tomadas con la aplicación).


En este caso, la regla más importante es que tanto el conjunto de validación como el de prueba deben ser lo más representativos posible de los datos que espera utilizar en producción, por lo que deben estar compuestos exclusivamente de imágenes representativas: puede mezclarlas y colocar la mitad en el conjunto de validación y la otra mitad en el de prueba (asegurándose de que no haya duplicados ni casi duplicados en ambos conjuntos). 

Después de entrenar su modelo con imágenes web, si observa que su rendimiento en el conjunto de validación es decepcionante, no sabrá si se debe a que su modelo se ha sobreajustado al conjunto de entrenamiento o si se debe simplemente a la discrepancia entre las imágenes web y las de la aplicación móvil. 

Una solución es conservar algunas imágenes de entrenamiento (de la web) en otro conjunto, que Andrew Ng denominó conjunto de entrenamiento-desarrollo (Figura 1-26). 

Una vez entrenado el modelo (en el conjunto de entrenamiento, no en el de entrenamiento-desarrollo), puede evaluarlo en este último. 

Si el modelo tiene un rendimiento deficiente, debe haber sobreajustado el conjunto de entrenamiento, por lo que debería intentar simplificarlo o regularizarlo, obtener más datos de entrenamiento y depurarlos. 

Pero si funciona bien en el conjunto de entrenamiento-desarrollo, puede evaluarlo en el conjunto de desarrollo. 

Si tiene un rendimiento deficiente, el problema debe provenir de la discrepancia de datos. 

Puedes intentar solucionar este problema preprocesando las imágenes web para que se parezcan más a las que tomará la aplicación móvil y luego reentrenando el modelo. 

Una vez que tengas un modelo que funcione bien tanto en el conjunto de entrenamiento y desarrollo como en el de desarrollo, puedes evaluarlo una última vez en el conjunto de pruebas para determinar su rendimiento probable en producción.

# Teorema de la "No hay nada gratis"


Un modelo es una representación simplificada de los datos. 

Las simplificaciones buscan descartar los detalles superfluos que probablemente no se generalicen a nuevas instancias. 

Al seleccionar un tipo de modelo en particular, se hacen suposiciones implícitas sobre los datos. 

Por ejemplo, si se elige un modelo lineal, se asume implícitamente que los datos son fundamentalmente lineales y que la distancia entre las instancias y la línea recta es simplemente ruido, que puede ignorarse con seguridad.


En un famoso artículo de 1996, David Wolpert demostró que si no se hace ninguna suposición sobre los datos, no hay razón para preferir un modelo sobre otro. 

Esto se denomina el teorema de la "No hay nada gratis" (NFL). 

Para algunos conjuntos de datos, el mejor modelo es un modelo lineal, mientras que para otros es una red neuronal.

No existe ningún modelo que garantice a priori un mejor funcionamiento (de ahí el nombre del teorema).


La ​​única manera de saber con certeza cuál es el mejor modelo es evaluarlos todos. 

Dado que esto no es posible, en la práctica se hacen suposiciones razonables sobre los datos y se evalúan solo unos pocos modelos razonables. 

Por ejemplo, para tareas sencillas, se pueden evaluar modelos lineales con varios niveles de regularización, y para un problema complejo, se pueden evaluar varias redes neuronales.

1. El aprendizaje automático consiste en construir sistemas que puedan aprender de los datos. Aprender significa mejorar en una tarea, dada una medida de rendimiento.


2. El aprendizaje automático es excelente para problemas complejos sin solución algorítmica, para reemplazar largas listas de reglas ajustadas manualmente, para construir sistemas que se adapten a entornos fluctuantes y, finalmente, para ayudar a los humanos a aprender (por ejemplo, minería de datos).


3. Un conjunto de entrenamiento etiquetado contiene la solución deseada (también conocida como etiqueta) para cada instancia.


4. Las dos tareas supervisadas más comunes son la regresión y la clasificación.


5. Las tareas no supervisadas comunes incluyen la agrupación en clústeres, la visualización, la reducción de dimensionalidad y el aprendizaje de reglas de asociación.


6. El aprendizaje por refuerzo probablemente funcione mejor si queremos que un robot aprenda a caminar en diversos terrenos desconocidos, ya que este es el tipo de problema que suele abordar. 

Podría ser posible expresar el problema como un problema de aprendizaje supervisado o semisupervisado, pero sería menos natural. 

7. Si no sabe cómo definir los grupos, puede usar un algoritmo de agrupamiento (aprendizaje no supervisado) para segmentar a sus clientes en grupos similares.

8. La detección de spam es un problema típico de aprendizaje supervisado: el algoritmo recibe muchos correos electrónicos junto con sus etiquetas (spam o no spam).


9. Un sistema de aprendizaje en línea puede aprender de forma incremental, a diferencia de un sistema de aprendizaje por lotes. Esto le permite adaptarse rápidamente tanto a datos cambiantes como a sistemas autónomos, y entrenarse con grandes cantidades de datos.


10. Los algoritmos fuera del núcleo pueden manejar grandes cantidades de datos que no caben en la memoria principal de un ordenador. 

Un algoritmo de aprendizaje fuera del núcleo divide los datos en minilotes y utiliza técnicas de aprendizaje en línea para aprender de ellos. 

11. Un sistema de aprendizaje basado en instancias memoriza los datos de entrenamiento; luego, al recibir una nueva instancia, utiliza una medida de similitud para encontrar las instancias aprendidas más similares y las utiliza para realizar predicciones.


12. Un modelo tiene uno o más parámetros que determinan qué predecirá dada una nueva instancia (p. ej., la pendiente de un modelo lineal). 

Un algoritmo de aprendizaje intenta encontrar valores óptimos para estos parámetros para que el modelo se generalice correctamente a nuevas instancias. 

Un hiperparámetro es un parámetro del propio algoritmo de aprendizaje, no del modelo (p. ej., el grado de regularización a aplicar).


13. Los algoritmos de aprendizaje basados ​​en modelos buscan un valor óptimo para los parámetros del modelo para que este se generalice correctamente a nuevas instancias.

Generalmente, entrenamos estos sistemas minimizando una función de costo que mide la ineficiencia del sistema al realizar predicciones con los datos de entrenamiento, además de una penalización por complejidad del modelo si este se regulariza. 

Para realizar predicciones, introducimos las características de la nueva instancia en la función de predicción del modelo, utilizando los valores de los parámetros encontrados por el algoritmo de aprendizaje.

Algunos de los principales desafíos del aprendizaje automático son la falta de datos, la mala calidad de los datos, la falta de representatividad de los datos, las características poco informativas, los modelos excesivamente simples que no se ajustan a los datos de entrenamiento y los modelos excesivamente complejos que los sobreajustan.


15. Si un modelo tiene un excelente rendimiento con los datos de entrenamiento, pero se generaliza deficientemente a nuevas instancias, es probable que esté sobreajustando los datos de entrenamiento (o que hayamos tenido mucha suerte con los datos de entrenamiento). 

Las posibles soluciones al sobreajuste son obtener más datos, simplificar el modelo (seleccionando un algoritmo más simple, reduciendo el número de parámetros o características utilizadas, o regularizando el modelo) o reducir el ruido en los datos de entrenamiento.


16. Un conjunto de prueba se utiliza para estimar el error de generalización que cometerá un modelo en nuevas instancias, antes de que se lance en producción.


17. Un conjunto de validación se utiliza para comparar modelos. 

Permite seleccionar el mejor modelo y ajustar los hiperparámetros. 

18. El conjunto de entrenamiento y desarrollo se utiliza cuando existe riesgo de discrepancia entre los datos de entrenamiento y los utilizados en los conjuntos de datos de validación y prueba (que siempre deben ser lo más parecidos posible a los datos utilizados una vez que el modelo esté en producción). 

El conjunto de entrenamiento y desarrollo es una parte del conjunto de entrenamiento que se reserva (el modelo no se entrena con él). 

El modelo se entrena con el resto del conjunto de entrenamiento y se evalúa tanto en el conjunto de entrenamiento y desarrollo como en el de validación. 

Si el modelo funciona bien en el conjunto de entrenamiento, pero no en el de entrenamiento y desarrollo, es probable que esté sobreajustando el conjunto de entrenamiento. 

Si funciona bien tanto en el conjunto de entrenamiento como en el de entrenamiento y desarrollo, pero no en el de validación, probablemente exista una discrepancia significativa entre los datos de entrenamiento y los datos de validación y prueba, y debería intentar mejorar los datos de entrenamiento para que se asemejen más a los de validación y prueba.
