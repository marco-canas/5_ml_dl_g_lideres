# Canalizaciones o Pipeline 


Una secuencia de componentes de procesamiento de datos se denomina canalización de datos. 

Las canalizaciones son muy comunes en los sistemas de aprendizaje automático, ya que hay muchos datos que manipular y muchas transformaciones que aplicar.


Los componentes suelen ejecutarse de forma asíncrona. 

Cada componente extrae una gran cantidad de datos, los procesa y genera el resultado en otro almacén de datos. 

Posteriormente, el siguiente componente de la canalización extrae estos datos y genera su propio resultado. 

Cada componente es bastante autónomo: la interfaz entre componentes es simplemente el almacén de datos. 

Esto facilita la comprensión del sistema (con la ayuda de un gráfico de flujo de datos) y permite que distintos equipos se centren en distintos componentes.

Además, si un componente falla, los componentes posteriores suelen poder seguir funcionando con normalidad (al menos durante un tiempo) utilizando únicamente la última salida del componente dañado. 

Esto confiere a la arquitectura una gran robustez.


Por otro lado, un componente dañado puede pasar desapercibido durante un tiempo si no se implementa una monitorización adecuada. 

Los datos se vuelven obsoletos y el rendimiento general del sistema disminuye.

# Seleccionar una medida de rendimiento


El siguiente paso es seleccionar una medida de rendimiento. 

Una medida de rendimiento típica para problemas de regresión es el error cuadrático medio (RMSE). 

Da una idea del grado de error que suele cometer el sistema en sus predicciones, otorgando mayor importancia a los errores grandes. 

La ecuación 2-1 muestra la fórmula matemática para calcular el RMSE.



Ecuación 2-1. Error cuadrático medio (RMSE)

$$
RMSE(X, y, h) = \sqrt{\frac{1}{m} \sum_{i = 1}^{m} \left( h\left(x^{(i)}\right) - y^{(i)} \right)^{2}}
$$

# Notaciones


Esta ecuación presenta varias notaciones de aprendizaje automático muy comunes que utilizaré a lo largo de este libro:



* $m$ es el número de instancias en el conjunto de datos en el que se mide el $RMSE$.

—Por ejemplo, si se evalúa el $RMSE$ en un conjunto de validación de $2000$ distritos, entonces $m = 2000.$

$x^{(i)}$ es un vector de todos los valores de las características (excluyendo la etiqueta) de la $i$-ésima instancia del conjunto de datos, e $y^{(i)}$ es su etiqueta (el valor de salida deseado para esa instancia). 

* $y$ es un vector que contiene las etiquetas de todas las instancias del conjunto de datos.



—Por ejemplo, si el primer distrito del conjunto de datos se encuentra en la longitud $-118,29^{o}$, latitud $33,91^{o}$, y tiene $1416$ habitantes con un ingreso medio de $\$38 372$, y el valor medio de la vivienda es de $\$156 400$ (ignorando otras características por ahora),
entonces:

$$
x^{(1)} = \begin{pmatrix}
-118.29 \\ 
33.91 \\ 
1,416 \\ 
38,372 \end{pmatrix}
$$

y $y^{(1)} = 156,400$.  

Para Geron, la coma es para indicar las unidades de mil. Mientras que el punto, es el punto decimal. 

* $\mathbf{X}$ es una matriz que contiene todos los valores de las características (excluyendo las etiquetas) de todas las instancias. Hay una fila por instancia, y la $i$-ésima fila es igual a la transpuesta de $x^{(i)}$, denotada $\left( x^{(i)} \right)^{T}$. 

Por ejemplo, si el primer distrito es como se acaba de describir, entonces la matriz $\mathbf{X}$ se verá así:  

$$
\mathbf{X} = \begin{pmatrix} 
\left( x^{(1)} \right)^{T} \\ 
\left( x^{(2)} \right)^{T}  \\ 
\vdots \\ 
\left( x^{(1999)} \right)^{T} \\ 
\left( x^{(2000)} \right)^{T}
\end{pmatrix} = \begin{pmatrix}  
-118.29 & 33.91 & 1,416 & 38,372 \\ 
\vdots & \vdots & \vdots & \vdots 
\end{pmatrix}
$$

* $h$ es tu función de predicción del sistema, también llamada hipótesis. Cuando se le proporciona a su sistema el vector de características de una instancia $x^{(i)}$, este genera un valor previsto $\hat{y}^{(i)} = h(x^{(i)})$ para esa instancia ($\hat{y}$ se pronuncia “y-hat”).
    - Por ejemplo, si su sistema predice que el precio medio de la vivienda en el primer distrito es de $\$158,400$, entonces $\hat{y}^{(1)} = h(x^{(1)}) = 158,400$. El error de predicción para este distrito es $\hat{y}^{(1)} – y^{(1)} = 2,000$.

* $RMSE(X,y,h)$ es la función de coste medida en el conjunto de ejemplos utilizando su hipótesis $h$.


Usamos letra cursiva minúscula para valores escalares ( como $m$ o $y^{(i)}$ ) y nombres de funciones (como $h$), letra negrita minúscula para vectores (como $\mathbf{x}^{(i)}$) y letra negrita mayúscula para matrices (como $\mathbf{X}$).

# El MAE 

Aunque el RMSE suele ser la medida de rendimiento preferida para las tareas de regresión, en algunos contextos puede ser preferible usar otra función, especialmente cuando hay muchos valores atípicos en los datos, ya que el RMSE es bastante sensible a ellos. 

En ese caso, puede considerar usar el error absoluto medio (MAE, también llamado desviación absoluta promedio), que se muestra en la Ecuación 2-2:

$$
MAE(\mathbf{X, y}, h) = \frac{1}{m} \sum_{i = 1}^{m} \left|h\left(x^{(i)} \right) - y^{(i)}  \right|
$$

Tanto el $RMSE$ como el $MAE$ son formas de medir la distancia entre dos vectores: el vector de predicciones y el vector de valores objetivo. 

Existen varias medidas de distancia, o normas, posibles:



• Calcular la raíz de una suma de cuadrados (RMSE) corresponde a la norma euclidiana: este es el concepto de distancia que todos conocemos.

También se denomina norma $ℓ_{2}$, denotada como $∥ \cdot ∥_{2}$ (o simplemente $∥ \cdot ∥$).



* Calcular la suma de absolutos (MAE) corresponde a la norma $ℓ_{1}$, denotada como $∥ \cdot ∥_{1}$. A veces se denomina norma de Manhattan porque mide la distancia entre dos puntos de una ciudad si solo se puede viajar a lo largo de manzanas ortogonales.


• De forma más general, la norma $ℓ_{k}$ de un vector $v$ con $n$ elementos se define como 

$$ ∥v∥_{k} = (|v_{1}
|^{k} + |v_{2}
|^{k} + \cdots + |v_{n}
|^{k})^{1/k}.   
$$ 

$ℓ_{0}$ indica el número de elementos distintos de cero en el vector, y $ℓ_{∞}$ indica el valor absoluto máximo del vector.  


Cuanto mayor sea el índice de norma, más se centra en valores grandes e ignora los pequeños. 

Por ello, el RMSE es más sensible a los valores atípicos que el MAE. 

Sin embargo, cuando los valores atípicos son exponencialmente escasos (como en una curva en forma de campana), el RMSE funciona muy bien y generalmente se prefiere.
