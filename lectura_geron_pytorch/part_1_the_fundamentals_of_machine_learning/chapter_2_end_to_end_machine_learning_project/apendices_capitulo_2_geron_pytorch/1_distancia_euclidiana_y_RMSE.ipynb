# La distancia euclidiana y RMSE

# Prompt  

Háblame de la distancia euclidiana que proviene de la norma $\ell_2$ y su utilidad en la medición del desempeño de un modelo de regresión en machine learning

$\ell_2$

La **distancia euclidiana** es una de las nociones geométricas más fundamentales en matemáticas y, a la vez, una de las más utilizadas en **machine learning**, especialmente cuando se evalúa el desempeño de **modelos de regresión**. 

Su origen está en la **norma $ \ell_2 $** de los espacios vectoriales euclidianos.

---



## 1. Norma $\ell_2$ y distancia euclidiana

Sea un vector $ \mathbf{x} = (x_1, x_2, \dots, x_n) \in \mathbb{R}^n $.
La **norma $ \ell_2 $** se define como:

$$
|\mathbf{x}|_2 ;=; \sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}.
$$



A partir de esta norma se induce la **distancia euclidiana** entre dos vectores
$ \mathbf{x}, \mathbf{y} \in \mathbb{R}^n $:

$$
d(\mathbf{x}, \mathbf{y})
;=; |\mathbf{x} - \mathbf{y}|_{2}
;=; \sqrt{\sum_{i=1}^n (x_i - y_i)^2}.
$$

Geométricamente, esta distancia representa la **longitud del segmento recto** que une los puntos $ \mathbf{x} $ y $ \mathbf{y} $ en el espacio euclidiano.

---



# 2. Interpretación en regresión

En un problema de regresión supervisada:

* $ \mathbf{y} = (y_1, \dots, y_n) $: valores reales (observados),
* $ \hat{\mathbf{y}} = (\hat{y}_1, \dots, \hat{y}_n) $: predicciones del modelo.

El **vector de errores o residuos** es:

$$
\mathbf{e} = \mathbf{y} - \hat{\mathbf{y}}.
$$



La norma $ \ell_2$ de este vector mide el **error total** del modelo:

$$
|\mathbf{e}|_2 = \sqrt{\sum_{i=1}^n (y_i - \hat{y}_i)^2}.
$$

Es decir, el desempeño del modelo se puede interpretar como la **distancia euclidiana entre el vector de valores reales y el vector de predicciones**.

---



# 3. Relación con métricas de desempeño

## a) Error cuadrático medio (MSE)

El **MSE** se define como:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2.
$$

Observa que:

$$
\text{MSE} = \frac{1}{n},|\mathbf{y} - \hat{\mathbf{y}}|_2^2.
$$

Minimizar el MSE equivale a **minimizar el cuadrado de la distancia euclidiana** entre $ \mathbf{y} $ y $ \hat{\mathbf{y}} $.

---



# b) Raíz del error cuadrático medio (RMSE)

$$
\text{RMSE} = \sqrt{\text{MSE}}
= \frac{1}{\sqrt{n}},|\mathbf{y} - \hat{\mathbf{y}}|_2.
$$

El RMSE es proporcional a la **distancia euclidiana promedio** entre observaciones y predicciones, y tiene la ventaja de estar en las **mismas unidades** de la variable respuesta.

---



# 4. Por qué la norma $ \ell_2 $ es tan usada en regresión

## 1. Interpretación geométrica clara

El aprendizaje consiste en **acercar** el vector de predicciones al vector real en el espacio $ \mathbb{R}^n $.



## 2. Propiedades analíticas favorables

* La función $ |\cdot|_2^2 $ es **convexa**.
* Es **derivable**, lo que facilita el uso de:

  * gradiente descendente,
  * métodos de optimización cerrados (como en regresión lineal clásica).



# 3. Conexión con mínimos cuadrados

La regresión lineal ordinaria busca:

$$
\min_{\boldsymbol{\theta}} ; |\mathbf{y} - X\boldsymbol{\theta}|_2^2,
$$

es decir, el **mejor ajuste en el sentido de mínimos cuadrados**, una proyección ortogonal del vector $\mathbf{y}$ sobre el subespacio generado por las columnas de $ X $.

---



# 5. Limitaciones de la distancia euclidiana

Aunque muy poderosa, la norma $ \ell_2 $:

* **Penaliza fuertemente los errores grandes** (outliers),
* Es sensible a la **escala** de las variables (requiere normalización),
* Puede no ser la mejor opción si el ruido no es aproximadamente gaussiano.

Por ello, en algunos contextos se prefieren normas como:

* $ \ell_1 $ (MAE, regresión robusta),
* pérdidas Huber o cuantílicas.

---



# 6. Resumen conceptual

* La **distancia euclidiana** proviene de la **norma $ \ell_2 $**.
* En regresión, mide cuán lejos están las predicciones de los valores reales.
* Métricas como **MSE y RMSE** son versiones normalizadas de esta distancia.
* Su éxito se debe a su interpretación geométrica, propiedades matemáticas y facilidad de optimización.
* Es la base teórica del criterio de **mínimos cuadrados** ampliamente usado en machine learning.

# 1. Interpretación geométrica → código con NumPy

Partimos de datos reales y predicciones como **vectores en $\mathbb{R}^n$**.



````python
import numpy as np # librería estrella en manejos vectoriales y matriciales 
````

````python

# Valores reales
y = np.array([3.0, 4.5, 6.0, 7.2, 8.1])
y 
````

````python

# Predicciones de un modelo
y_hat = np.array([2.8, 4.9, 5.5, 7.0, 8.6])
y_hat 
````

````python

# Vector de errores
e = y - y_hat
e
````

````python
np.sum(e)
````

# Norma ℓ2 (distancia euclidiana)



````python
l2_distance = np.linalg.norm(e, ord=2) # el parámetro ord me define que tipo de norma estoy utilizando 
l2_distance
````


Esto implementa directamente:

$$
|\mathbf{y} - \hat{\mathbf{y}}|_2
$$



**Lectura didáctica**:

> “Esta cantidad es la longitud del vector error; mide qué tan lejos está el modelo de la realidad en promedio geométrico”.

---



### Norma ℓ1 (distancia Manhattan)

```python


````python
l1_distance = np.linalg.norm(e, ord=1) # ord en 1 implementa la norma l1
l1_distance
````


$$
|\mathbf{y} - \hat{\mathbf{y}}|_1 = \sum |y_i - \hat{y}_i|
$$



➡️ **Lectura didáctica**:

> “Aquí el error total se mide como la suma de los desvíos absolutos, sin exagerar los errores grandes”.

---

# 2. Conexión directa con métricas de desempeño

```python


````python
n = len(y)

MSE = np.mean(e**2)
RMSE = np.sqrt(MSE)
MAE = np.mean(np.abs(e))

MSE, RMSE, MAE
````


Relación clave para los estudiantes:

$$
\text{RMSE} = \frac{1}{\sqrt{n}} |\mathbf{y} - \hat{\mathbf{y}}|_2
\quad \text{y} \quad
\text{MAE} = \frac{1}{n} |\mathbf{y} - \hat{\mathbf{y}}|_1
$$

# 3. Visualización geométrica simple con Matplotlib (caso $n = 2$)

Este ejemplo es **muy potente en clase**, porque permite “ver” las normas.


````python
import matplotlib.pyplot as plt

# Punto real y predicción en R^2
y = np.array([4, 6])
y_hat = np.array([6, 3])

error = y - y_hat

plt.figure()
plt.quiver(0, 0, error[0], error[1], angles='xy', scale_units='xy', scale=1, label = 'vector e de error')
plt.scatter(*y, label="y real")
plt.scatter(*y_hat, label="y predicho")
plt.scatter(0, 0)

plt.axhline(0)
plt.axvline(0)
plt.grid(True)
plt.axis([-4,7,0, 7])
plt.legend()
plt.title("Vector error como distancia geométrica")
plt.savefig(r"C:\Users\marco\Downloads\vector_observaciones_vs_vector_predicciones.png")
plt.show()
````

**Mensaje pedagógico**:

* Norma $ \ell_2$: longitud recta del vector.
* Norma $\ell_1$: recorrido “en escalera” horizontal + vertical.


````python
# Visualización de la distancia inducida por la norma L1 (distancia Manhattan)
import matplotlib.pyplot as plt

# Definimos dos puntos en el plano
x1, y1 = 1, 2
x2, y2 = 5, 6

# Cálculo de la distancia L1
dist_l1 = abs(x2 - x1) + abs(y2 - y1)

# Creamos la figura
plt.figure()
plt.scatter([x1, x2], [y1, y2])
plt.plot([x1, x2], [y1, y1])  # desplazamiento horizontal
plt.plot([x2, x2], [y1, y2])  # desplazamiento vertical

# Anotaciones
plt.text(x1, y1, " A(x1, y1)", fontsize=10)
plt.text(x2, y2, " B(x2, y2)", fontsize=10)
plt.text((x1 + x2) / 2, y1 - 0.3, "|x2 - x1|", ha='center')
plt.text(x2 + 0.1, (y1 + y2) / 2, "|y2 - y1|", va='center')

plt.title(f"Distancia inducida por la norma L1 (d₁ = {dist_l1})")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(True)
plt.axis('equal')
plt.savefig(r"C:\Users\marco\Downloads\interpretacion_distancia_inducida_por_l1.png")
plt.show()
````


# 4. Comparación ℓ2 vs ℓ1 con Scikit-Learn

## a) Regresión con pérdida ℓ2 (MSE): LinearRegression


````python
%%time 
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([3, 4, 6, 7, 8])

model_l2 = LinearRegression()
model_l2.fit(X, y)

y_hat_l2 = model_l2.predict(X)
y_hat_l2
````

# Optimiza:

$$
\min |\mathbf{y} - \hat{\mathbf{y}}|_2^2
$$

---



## b) Regresión con pérdida ℓ1 (MAE): LAD / Quantile Regression

```python


````python
from sklearn.linear_model import QuantileRegressor

model_l1 = QuantileRegressor(quantile=0.5, alpha=0)
model_l1.fit(X, y)

y_hat_l1 = model_l1.predict(X)
y_hat_l1
````

# Optimiza:

$$
\min |\mathbf{y} - \hat{\mathbf{y}}|_1
$$

---



# 5. Visualización comparativa de ambos ajustes


````python
plt.figure()
plt.scatter(X, y, label="Datos reales", color = 'red')          # visualiza los datos 
plt.plot(X, y_hat_l2, label="Regresión ℓ2 (MSE)")
plt.plot(X, y_hat_l1, label="Regresión ℓ1 (MAE)")
plt.legend()
plt.grid(True)
plt.title("Comparación geométrica: norma ℓ2 vs norma ℓ1")
plt.savefig(r"C:\Users\marco\Downloads\Comparacion_entre_modelos_dependiendo_de_la_norma.png")
plt.show()
````

# Estado del arte



El interés por integrar la tecnología en la enseñanza de las matemáticas tiene raíces en las propuestas de **Papert (1980, 1996)**, quien con su enfoque constructionista concibió la computadora como un “objeto para pensar”, capaz de mediar la construcción activa de significados matemáticos mediante la programación. Esta perspectiva se enlaza con los planteamientos de **Piaget (1970)** y **Vygotsky (1978)**, quienes reconocieron la importancia de la experimentación y de las herramientas culturales en la formación del pensamiento. Bajo esta mirada, el uso de entornos digitales no constituye una amenaza para la enseñanza tradicional de las matemáticas, sino un recurso que amplía las posibilidades de aprendizaje.



A comienzos del siglo XXI, la noción de **pensamiento computacional** introducida por **Wing (2006)** renovó el debate al proponer que esta competencia debía enseñarse de manera transversal desde los niveles iniciales hasta la educación superior, situándola al mismo nivel que la lectura, la escritura y las matemáticas. 

Este enfoque se alinea con los planteamientos del **NCTM (2000, 2014)**, que destacó la tecnología como un componente esencial para la experimentación, la representación y la comunicación en matemáticas. 

En paralelo, aportes teóricos como la **teoría de registros de representación semiótica de Duval (1993, 2006)** y el **Enfoque Ontosemiótico de Godino (2007, 2017)** muestran que el aprendizaje matemático se enriquece al transitar entre múltiples sistemas de representación, incluyendo los digitales.



En el contexto actual, organizaciones como la **ACM** y el **IEEE** insisten en que las competencias computacionales son indispensables en la formación profesional, lo que refuerza la pertinencia de integrar lenguajes como **Python, LaTeX, Markdown e inglés** en ambientes como **Jupyter Notebooks** para la enseñanza universitaria de las matemáticas. 

Así, el estado del arte evidencia un consenso creciente: lejos de ser un obstáculo, el pensamiento computacional y el uso de lenguajes digitales constituyen una vía para fortalecer la comprensión matemática, formar profesionales más competentes y responder a las demandas del mundo laboral y científico contemporáneo.



# Estado del arte con citas



El uso de la tecnología en la enseñanza de las matemáticas encuentra uno de sus pilares en **Papert (1980)**, quien afirmaba que *“los niños no aprenden sobre la computadora, aprenden con la computadora”* (p. 6). 

En su visión, la programación se convierte en un medio para la construcción activa de significados matemáticos, idea que se enlaza con la epistemología genética de **Piaget (1970)**, quien sostuvo que *“el conocimiento se construye a través de la acción”* (p. 15), y con la perspectiva de **Vygotsky (1978)**, al señalar que *“toda función en el desarrollo cultural del niño aparece dos veces: primero en el plano social y después en el individual”* (p. 57). 

Estos enfoques coinciden en otorgar a las herramientas culturales un papel esencial en la mediación del aprendizaje.



Posteriormente, **Wing (2006)** acuñó el término pensamiento computacional, definiéndolo como *“una habilidad fundamental para todos, no solo para los científicos de la computación”* (p. 33). 

Esta autora subrayó que debía enseñarse de manera transversal, en paralelo con la lectura, la escritura y la matemática. En la misma línea, el **NCTM (2000)** destacó que *“la tecnología es esencial en la enseñanza y el aprendizaje de las matemáticas; influye en la matemática que se enseña y mejora el aprendizaje de los estudiantes”* (p. 24).



Desde la didáctica de las matemáticas, **Duval (2006)** mostró que *“la comprensión matemática requiere la coordinación de al menos dos registros de representación diferentes”* (p. 107), lo que legitima la incorporación de registros digitales como LaTeX, Markdown y Python. 

Por su parte, **Godino (2017)** planteó que *“el significado de un objeto matemático está constituido por el sistema de prácticas en el que interviene”* (p. 7), con lo cual la práctica digital se convierte en parte constitutiva del aprendizaje matemático.



Finalmente, organismos internacionales como la **ACM y el IEEE (2013)** enfatizan que *“las habilidades computacionales son esenciales para el ejercicio profesional en cualquier disciplina”* (p. 12). 

De esta manera, se consolida un consenso: la enseñanza universitaria de las matemáticas debe integrar herramientas contemporáneas como **Jupyter Notebooks, Python, LaTeX, Markdown e inglés**, no para competir con la enseñanza tradicional, sino para **fortalecer la comprensión matemática, potenciar la comunicación científica y responder a las demandas del mundo laboral**.
