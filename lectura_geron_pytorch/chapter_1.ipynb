# Objetivo y enfoque


Este libro asume que no sabes prácticamente nada sobre aprendizaje automático. 

Su objetivo es brindarte los conceptos, las herramientas y la intuición necesarios para implementar programas capaces de aprender de los datos.


Abarcaremos una gran variedad de técnicas, desde las más sencillas y comunes (como la regresión lineal) hasta algunas de las técnicas de aprendizaje profundo que suelen ganar concursos. 

Para ello, utilizaremos Python, el lenguaje líder en ciencia de datos y aprendizaje automático, así como frameworks de Python de código abierto y listos para producción.

* Scikit-Learn es muy fácil de usar, e implementa numerosos algoritmos de aprendizaje automático de forma eficiente, lo que lo convierte en una excelente puerta de entrada para el aprendizaje automático. 

Fue creado por David Cournapeau en 2007, y posteriormente dirigido por un equipo de investigadores del Instituto Francés de Investigación en Informática y Automatización (Inria), y recientemente por Probabl.ai.


* PyTorch es una biblioteca potente y flexible para el aprendizaje profundo. Permite entrenar y ejecutar todo tipo de redes neuronales de forma eficiente y puede distribuir los cálculos entre múltiples GPU (unidades de procesamiento gráfico).


PyTorch (PT) fue desarrollado por el laboratorio de investigación de IA de Facebook (FAIR) y se lanzó por primera vez en 2016. Evolucionó a partir de Torch, un framework más antiguo codificado en Lua. 

En 2022, PyTorch se transfirió a la Fundación PyTorch, bajo la Fundación Linux, para promover el desarrollo impulsado por la comunidad.
También utilizaremos estas bibliotecas de aprendizaje automático de código abierto en el proceso:

* XGBoost en el Capítulo 6 para implementar una potente técnica llamada gradient boosting.


Bibliotecas Hugging Face en los Capítulos 13 y 15 para descargar conjuntos de datos y modelos preentrenados, incluyendo modelos de transformadores. Los transformadores son increíblemente potentes y versátiles, y son el componente principal de prácticamente todos los asistentes de IA actuales.


* Gymnasium en el Capítulo 19 para aprendizaje por refuerzo (es decir, entrenamiento de agentes autónomos).


El libro prioriza un enfoque práctico, desarrollando una comprensión intuitiva del aprendizaje automático a través de ejemplos prácticos concretos y un poco de teoría.

# Hoja de ruta
Este libro está organizado en dos partes. La Parte I, "Fundamentos del Aprendizaje Automático", abarca los siguientes temas:  

* Qué es el aprendizaje automático, qué problemas intenta resolver y las principales categorías y conceptos fundamentales de sus sistemas.
* Los pasos de un proyecto típico de aprendizaje automático.
* Aprendizaje mediante el ajuste de un modelo a los datos.

* Minimización de una función de coste (es decir, una medida de errores de predicción).
* Manejo, limpieza y preparación de datos.
* Selección e ingeniería de características (es decir, campos de datos).
* Selección de un modelo y ajuste de hiperparámetros mediante validación cruzada (p. ej., entrenamiento con múltiples variantes del modelo y elección de la que mejor se adapta a los datos que no se observaron durante el entrenamiento).
* Los desafíos del aprendizaje automático, en particular el subajuste y el sobreajuste (el equilibrio entre sesgo y varianza).
* Los algoritmos de aprendizaje más comunes: regresión lineal y polinómica, regresión logística, k-vecinos más cercanos, árboles de decisión, bosques aleatorios y métodos de conjunto.
* Reducción de la dimensionalidad de los datos de entrenamiento para combatir la “maldición de dimensionalidad”
* Otras técnicas de aprendizaje no supervisado, como la agrupación en clústeres, la estimación de densidad y la detección de anomalías

# La Parte II, “Redes neuronales y aprendizaje profundo”, abarca los siguientes temas:  

* Qué son las redes neuronales y para qué sirven


* Construcción y entrenamiento de redes neuronales profundas con PyTorch


* Las arquitecturas de redes neuronales más importantes: redes neuronales de propagación hacia adelante para datos tabulares; redes convolucionales para visión artificial; redes recurrentes y redes de memoria a corto plazo (LSTM) para procesamiento de secuencias; codificadores, decodificadores, transformadores, modelos de espacio de estados (SSM) y arquitecturas híbridas para procesamiento del lenguaje natural, visión y más; autocodificadores, redes generativas antagónicas (GAN) y modelos de difusión para aprendizaje generativo. 


* Cómo construir un agente (por ejemplo, un bot en un juego) que pueda aprender buenas estrategias mediante ensayo y error, utilizando aprendizaje por refuerzo
* Carga y preprocesamiento eficiente de grandes cantidades de datos
* La primera parte se basa principalmente en Scikit-Learn; la segunda parte utiliza principalmente PyTorch

# Cambios entre las versiones de TensorFlow y PyTorch


Escribí tres ediciones de TensorFlow (TF) de este libro, publicado por O'Reilly en 2017, 2019 y 2022. 

TF fue la biblioteca líder de aprendizaje profundo durante muchos años, utilizada internamente por Google y, por lo tanto, optimizada para la producción a escala. 

Sin embargo, PyTorch ha tomado gradualmente la delantera gracias a su simplicidad, flexibilidad y apertura: ahora domina los artículos de investigación y los proyectos de código abierto, lo que significa que la mayoría de los nuevos modelos están disponibles primero en PyTorch. 

Como resultado, la industria también se ha inclinado gradualmente hacia PyTorch.

En los últimos años, Google ha reducido sus inversiones en TensorFlow y se ha centrado más en JAX, otra excelente biblioteca de aprendizaje profundo con una excelente combinación de cualidades tanto para investigación como para producción. 

Sin embargo, su adopción sigue siendo baja en comparación con PyTorch.


¡Por eso elegí usar PyTorch esta vez! O’Reilly y yo decidimos que fuera la primera edición de una nueva serie de PyTorch en lugar de la cuarta edición de la serie original.
